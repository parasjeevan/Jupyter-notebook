```python
#10.How many sharks participated in this show and What were their names?
shark_names = [col.replace('_present', '').capitalize() for col in STI_data.columns if col.endswith('_present')]
print(shark_names)
```

```python
#11.Find appearance of each sharks? 

SA = {}
for column_name in STI_data.columns:
    if "_present" in column_name:
        shark_name = column_name.replace("_present", "").capitalize()
        appearances = STI_data[column_name].sum()  
        SA[shark_name] = appearances
        
SA = pd.DataFrame(list(SA.items()), columns=['Shark_Name', 'Appearance_Count'])
SA
        
```

```python
#13.Find the total number of amount invested in this show? 
total_investment = STI_data['deal_amount'].sum()
total_investment
```

```python
#14.Find the equity percent that each sharks gets?

sharks = ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal','peyush_deal','ghazal_deal']

shark_equity_total = {}

for column in sharks:
    sharks = column.replace('_deal', '').capitalize()
    total_equity_for_shark = STI_data[column].sum()
    shark_equity_total[sharks] = total_equity_for_shark
total_equity_offered = STI_data['equity_per_shark'].sum()

shark_equity_percent = {}

for shark, total_equity in shark_equity_total.items():
    equity_percent = (total_equity / total_equity_offered) * 100
    shark_equity_percent[shark] = equity_percent
    
shark_equity_percent
```

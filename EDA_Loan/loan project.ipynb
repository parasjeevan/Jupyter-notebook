# 5. Impute null values with Mode in Gender Column

```python
loan_data = pd.read_csv(r"J:\@IOT Academy@\Mini_Projects\03_EDA_Loan\loan_sanction.csv")
# Calculate the mode of the 'Gender' column
gender_mode = loan_data['Gender'].mode()[0]

# Replace null values with the mode
loan_data['Gender'].fillna(gender_mode, inplace=True)
```

# 7. Impute null values with Median in LoanAmount Column
